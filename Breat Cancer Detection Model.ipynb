```python
# import libraries
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline

import warnings
warnings.filterwarnings('always')
warnings.filterwarnings('ignore')
```

```python
# getting the dataset
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
data = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)  
# data is variable where we are storing. here data is breast cancer
```

```python
data['class'] = breast_cancer.target
#adding target to this class cloumn. class here is lable i.e 0 or 1
```

```python
data.head(10)
```

```python
# print shape of the dataset

print('The shape of the dataset: ',data.shape)
```

```python
# count plot of class

sns.countplot(x='class',data=data)
```

```python
# data visualization:

#correlation matrix

plt.figure(figsize=(20,10))
sns.heatmap(data.corr(),annot=True)
```

```python
x=data.drop('class',axis=1)
y=data['class']
```

```python
# implement logistics regression

# train test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# print x_train
x_train
```

```python
# print y_train
y_train
```

```python
#import LogisticRegression from sklearn
from sklearn.linear_model import LogisticRegression
#loading LogisticRegression to classifier
classifier = LogisticRegression()
#training the model. X is traing data. Y is test data
classifier.fit(x_train, y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
#import accuracy_score
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
# classification report
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
# roc curve
from sklearn import metrics
metrics.plot_roc_curve(classifier, x_test, y_test)
plt.show()
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
confusion=confusion_matrix(y_test,y_pred)
confusion
```
